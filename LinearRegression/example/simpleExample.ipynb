```python
from LinearRegression import LinearRegression
```

```python
def importData():
    import pickle
    # function to import and export data from cPickle format
    def unpickle(file):
        with open(file, 'rb') as fo:
            dict = pickle.load(fo, encoding='bytes')
        return dict
    
    prefix = 'C:/Users/peill/Documents/Sigma_Clermont/MS_ESD/M5/APYT_TP_GradientDescent_'
    featuresFile = prefix + 'Sigma_features.pkl'
    targetFile = prefix + 'Sigma_labels.pkl'
    
    features = unpickle(featuresFile)
    target = unpickle(targetFile)
    m = target.shape[0]
    print(m)
    target = target.reshape(-1,1)
    print(target.shape)
    
    return features, target
```

```python
features, target = importData()
```

```python
model = LinearRegression(X=features, y_gt=target, lr=3e-9, eps=1e-5)
```

```python
model.fit()
```

```python
print(model.score())
model.plot_featureimportance()
```

```python
help(model.plot_featureimportance)
```

# <b>SIMPLE EXAMPLE</b>

```python
import numpy as np
import pandas as pd
x1, x2 = np.linspace(0,1), np.linspace(0,1)
noise = np.random.random(50)
y = 2*x1 + 3*x2 + noise*2 + 5
df = pd.DataFrame({'x1':x1, 'x2':x2,'y':y})
noise
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
X, y = df[['x1','x2']].to_numpy(), df[['y']].to_numpy()
model = LinearRegression(X=X, y_gt=y, lr=0.1, eps=1e-90)
model.fit()
print(f'R-squared: {model.score():.04f}')
```

```python
model.plot_featureimportance(names=['x1','x2'])
```
