# <b>TOY EXAMPLE</b>

```python
from LinearRegression import LinearRegression
```

```python
import numpy as np
import pandas as pd
x1, x2 = np.linspace(0,1), np.linspace(0,1)
noise = np.random.random(50)
y = 2*x1 + 3*x2 + noise*2 + 5
df = pd.DataFrame({'x1':x1, 'x2':x2,'y':y})
noise
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
X, y = df[['x1','x2']].to_numpy(), df[['y']].to_numpy()
```

```python
model = LinearRegression(X=X, y_gt=y, lr=0.1, eps=1e-9)
model.fit()
model.score()
model.plot_featureimportance(names=np.array(['x1','x2']))
```

```python
model.params
```

## Normalization

```python
model_norm = LinearRegression(X=X, y_gt=y, lr=0.1, eps=1e-9, std=True)
model.fit()
model.score()
model.plot_featureimportance(names=np.array(['x1','x2']))
```
